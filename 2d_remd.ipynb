# Model System with Orthogonal Barrier

We are going to do umbrella sampling on a simple 2D model system, where there is a barrier orthogonal to the umbrella sampling reaction coordinate.

```python
width = 200
height = 800

def make_energy(barrier_height, spring_y, spring_k):
    energy = np.zeros((width, height)) + 1e100
    energy[25:75, 25:450] = 0
    energy[125:175, 350:775] = 0
    energy[75:125, 350:450] = barrier_height
    spring_energy = np.linspace(0, height - 1, height)
    spring_energy = 0.5 * spring_k * (spring_y - spring_energy) ** 2
    energy += spring_energy
    return energy

for i in range(9):
    pp.subplot(1, 9, i+1)
    x = i * 100.
    energy = make_energy(200, x, 0.04)
    pp.imshow(energy.T, vmax=250)
```

We are doing umbrella sampling along the y-axis, but there is a barrier along the x-axis at y=400. This will cause problems with REMD efficiency.

# Cython

We're going to define a big block of cython code. This is a bit gross looking, but is ~12x faster than pure python code.

We'll be defining three things:

1. Walkers
2. States
3. The Ladder

The ladder will have `n_steps steps`. Each step `i` has `n_reps[i]` copies.

```python
%%cython

import random
import math
cdef extern from "math.h":
    double exp(double)
import numpy as np
cimport numpy as np
from collections import namedtuple

# setup the c random number generator so we don't have to call python all the time
from libc.stdlib cimport rand, srand, RAND_MAX
# set the random seed
cdef int rand_range(int lower, int upper):
    delta = upper - lower
    cdef int r = rand() % (delta + 1) + lower
    return r



cdef class Walker(object):
    cdef public int x, y
    cdef int width, height
    cdef double[:, ::1] energy_landscape
    
    def __init__(self, np.ndarray[np.double_t, ndim=2] energy_landscape, int x, int y, int width, int height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.energy_landscape = energy_landscape
    
    def steps(self, int n_steps):
        cdef int i
        cdef int delta_x
        cdef int delta_y
        cdef int proposed_x
        cdef int proposed_y
        cdef double delta_e
        cdef int accept
        for i in range(n_steps):
            #delta_x = random.choice([-3, -2, -1, 0, 1, 2, 3])
            #delta_y = random.choice([-3, -2, -1, 0, 1, 2, 3])
            delta_x = rand_range(-3, 3)
            delta_y = rand_range(-3, 3)
            proposed_x = self.x + delta_x
            proposed_y = self.y + delta_y
            if proposed_x < 0 or proposed_x >= self.width:
                return self.x, self.y
            if proposed_y < 0 or proposed_y >= self.height:
                return self.x, self.y
        
            delta_e = self.proposed_energy(proposed_x, proposed_y) - self.proposed_energy(self.x, self.y)
            accept = False
            if delta_e < 0:
                accept = True
            else:
                metrop = exp(-delta_e)
                if random.random() < metrop:
                    accept = True
                else:
                    accept = False
        
            if accept:
                self.x = proposed_x
                self.y = proposed_y
                
        return self.x, self.y
    
    cdef double proposed_energy(self, int x, int y):
        return self.energy_landscape[x, y]
    
    def energy(self):
        return self.energy_landscape[self.x, self.y]


class State(object):
    def __init__(self, init_state):
        self._curr_state = init_state
        self._history = [init_state]
    
    @property
    def curr_state(self):
        return self._curr_state
    
    @curr_state.setter
    def curr_state(self, new_state):
        self._history.append(new_state)
        self._curr_state = new_state
    
    @property
    def history(self):
        return self._history
    
    
UmbrellaParam = namedtuple('UmbrellaParam', 'center force_k')

cdef class Ladder(object):
    cdef public int n_steps, n_walkers, round_trips
    cdef public list states, walker_history
    cdef list step_params, walkers, state_indices, state_walker
    cdef list walker_permute 
    cdef public double[:] tried, accepted, upward, downward, n_up, n_down

    def __init__(self, step_params, step_states, energy_func, width, height):
        '''
        step_params is a list of UmbrellaParam objects, one for each step
        
        step_states is a list of lists, where is a list of one or more starting states for each step.
            # some constraints
            len(step_states) == len(step_params)
            for i in range(len(step_states):
                len(step_states[i]) > 0
        '''
        self.n_steps = len(step_params)
        self.step_params = step_params
        
        self.states = []
        self.walkers = []
        self.state_indices = [0]
        state_counter = 0
        self.state_walker = []
        for index, step in enumerate(step_states):
            for state in step:
                state_counter += 1
                self.states.append(state)
                parms = self.step_params[index]
                new_walker = Walker(energy_func(parms.center, parms.force_k),
                                    state.curr_state[0], state.curr_state[1],
                                    width, height)
                self.walkers.append(new_walker)
                self.state_walker.append(index)
            self.state_indices.append(state_counter)
        
        self.n_walkers = len(self.walkers)
        self.walker_permute = range(self.n_walkers)
        self.walker_history = []
        self.walker_history.append(self.state_walker)
        self.tried = np.zeros(self.n_steps - 1)
        self.accepted = np.zeros(self.n_steps - 1)
        self.upward = np.zeros(self.n_walkers)
        self.downward = np.zeros(self.n_walkers)
        self.n_up = np.zeros(self.n_steps)
        self.n_down = np.zeros(self.n_steps)
        self.round_trips = 0
    
    def update(self, int n_updates):
        cdef int i, j, x_i, x_j, y_i, y_j, offset, index_i, index_j
        cdef list lower_pairs
        cdef Walker walker_i, walker_j
        for i in range(n_updates):
            # do a mc update of each walker
            for walker in self.walkers:
                walker.steps(10)
        
            # do some REMD moves
            for trial in range(1):
                offset = random.choice([0, 1])
                lower_pairs = [i for i in range(offset, self.n_steps - 1, 2)]
                for i in lower_pairs:
                    j = i + 1
                    self.tried[i] += 1.0
                
                    index_i = rand_range(self.state_indices[i], self.state_indices[i+1] - 1)
                    index_j = rand_range(self.state_indices[j], self.state_indices[j+1] - 1)
                    walker_i = self.walkers[index_i]
                    walker_j = self.walkers[index_j]
                    x_i = walker_i.x
                    y_i = walker_i.y
                    x_j = walker_j.x
                    y_j = walker_j.y
                            
                    delta = (walker_i.proposed_energy(x_j, y_j) + walker_j.proposed_energy(x_i, y_i)
                             - (walker_i.proposed_energy(x_i, y_i) + walker_j.proposed_energy(x_j, y_j)))
                    if delta < 0:
                        swap = True
                    else:
                        factor = exp(-delta)
                        if random.random() < factor:
                            swap = True
                        else:
                            swap = False
                    
                    if swap:
                        (walker_i.x, walker_j.x) = (walker_j.x, walker_i.x)
                        (walker_i.y, walker_j.y) = (walker_j.y, walker_i.y)
                        (self.walker_permute[index_i], self.walker_permute[index_j]) = (self.walker_permute[index_j],
                                                                                        self.walker_permute[index_i])
                        self.accepted[i] += 1.0
        
            # update the states to keep track of history
            rep_index = []
            for state, walker in zip(self.states, range(len(self.walkers))):
                index = self.walker_permute.index(walker)
                w = self.walkers[index]
                state.curr_state = (w.x, w.y)
                rep_index.append(self.state_walker[index])
            
                # keep track of f_up curve
                if self.state_walker[index] == 0:
                    if self.downward[walker] == 1:
                        self.round_trips += 1
                    self.upward[walker] = 1
                    self.downward[walker] = 0
                if self.state_walker[index] == (self.n_steps - 1):
                    self.upward[walker] = 0
                    self.downward[walker] = 1
                self.n_up[self.state_walker[index]] += self.upward[walker]
                self.n_down[self.state_walker[index]] += self.downward[walker]
        
            self.walker_history.append(rep_index)
        
    cpdef compute_acceptance(self):
        cdef int i
        output = np.zeros(self.n_steps - 1)
        for i in range(self.n_steps - 1):
            output[i] = self.accepted[i] / self.tried[i]
        return output
    
    cpdef compute_fup(self):
        cdef int i
        output = np.zeros(self.n_steps)
        for i in range(self.n_steps):
            output[i] = self.n_up[i] / (self.n_up[i] + self.n_down[i])
        return output
```

# Without Barriers

Now we'll run an umbrella sampling / Hamiltonian REMD simulation where there is no barrier along the x-axis. We'll use 24 replicas with umbrella potentials spaced evenly between y=100 and y=700.

```python
centers = np.linspace(100, 700, 24)
parms = [UmbrellaParam(x, 0.001) for x in centers]

def num_copies(center):
    if center >= 300 and center <= 500:
        return 1
    else:
        return 1
    
states = [[State((50 if c < 400 else 150, c)) for i in range(num_copies(c))] for c in centers]
ladder1 = Ladder(parms, states, functools.partial(make_energy, 0.0), width, height)
```

Let's generate some samples. The Cython voodoo gets from 10 minutes to about 50 seconds for this.

```python
%time ladder1.update(500000)
```

## Trajectory

```python
def plot_trajectories(ladder):
    pp.figure(figsize=(12,12))
    for index, state in enumerate(ladder.states):
        pp.subplot(math.ceil(len(ladder.states) / 4.0), 4.0, index + 1)
        h = np.array(state.history)
        pp.plot(h[:, 0], h[:, 1], 'k-')
        pp.xlim(0, 200)
        pp.ylim(0, 800)
        
plot_trajectories(ladder1)
```

Each replica samples the same distribution, which indicates that we've achived equilibrium sampling. Replicas are able to easily cross the "barrier" region, because in this case it's not a barrier.

## Replica Walk

Let's plot the replica index of each structure as it traverses the ladder.

```python
def plot_walks(ladder):
    hist = np.array(ladder.walker_history).T
    pp.figure(figsize=(18, 12))
    for i in range(ladder.n_walkers):
        pp.subplot(math.ceil(ladder.n_walkers / 4.0), 4, i + 1)
        pp.plot(hist[i,:], color='black')

plot_walks(ladder1)
```

Overall, the replicas are able traverse the ladder efficiently. The behavior looks similar to a 1D random walk with a reflecting boundary. There are no signs of obvious bottlenecks.

## Exchange Rates

```python
def plot_accept(ladder, color='k-o'):
    pp.plot(ladder.compute_acceptance(), color);

    
plot_accept(ladder1)
```

The exchange rate is relatively uniform. There are some slight blips due to there not being exactly the same amount of overlap between replicas due to the presence of the barrier region.

## The $f_{up}$ Curve

```python
def plot_fup(ladder, color='k-o'):
    pp.plot(ladder.compute_fup(), color)
    pp.plot([0, ladder.n_steps], [1.0, 0.0], 'k--')

plot_fup(ladder1)
```

The $f_{up}$ curve is close to linear. The little blips it has are due to the non-uniformity of the exchange rate.

## Round Trips

```python
print ladder1.round_trips
```

Our aim is to maximize the number of statistically independent samples at the lowest temperature. In the case of MELD, what we want to maximize is the number of independent "folding" trajectories. These begin at the highest temperature and end up at the lowest temperature. We call these _round trips_.

**Note: it's actually not entirely clear what we are trying to opitmize, but this seems like a convenient proxy.**

# With a  barrier

In this case, we're going to increase the barrier from 0 to 5 kT, so that there is actually a substantial barrier orthogonal to our reaction coordinate. The expectation is that this should slow down dynamics and lead to poor mixing of the REMD chain--and ultimately to fewer round trips.

```python
centers = np.linspace(100, 700, 24)
parms = [UmbrellaParam(x, 0.001) for x in centers]

def num_copies(center):
    if center >= 300 and center <= 500:
        return 1
    else:
        return 1
    
states = [[State((50 if c < 400 else 150, c)) for i in range(num_copies(c))] for c in centers]
ladder2 = Ladder(parms, states, functools.partial(make_energy, 5.0), width, height)
```

```python
ladder2.update(500000)
```

## Trajectory

```python
plot_trajectories(ladder2)
```

As expected, the density in the barrier region is reduced and there are fewer barrier crossings.

## Replica Walk

```python
plot_walks(ladder2)
```

There is now a clear bottleneck around y=400 or replica ~11. Replicas tend to bounce off of this region and crossings are rare.

## Exchange Rate

```python
plot_accept(ladder1, 'k-o')
plot_accept(ladder2, 'r-o')
```

The exchange rate is very similar to the system without the orthogonal barrier. The differences are due to the fact that the barrier region has lower probability, which changes the overlap slightly.

**This is critically important.** The exchange rate does not depend on the kinetics of the replica ladder. The exchange rate is a static thermodynamic property due to the "overlap" between adjacent replicas. The exchange rate affects the kinetics, but **the kinetics does not affect the exchange rate**. This is why much of literature on replica exchange is misleading--it begins fromt the assumption that optmizing exchange rates is the way to optimize REMD.

## The $f_{up}$ Curve

```python
plot_fup(ladder1, 'k-o')
plot_fup(ladder2, 'r-o')
```

The $f_{up}$ curve is not linear. This indicates that there is a barier around replicas 5--15. This barrier in mixing is, of course, due to the energy barrier along the x coordinate.

In the scheme of Hansmann, we would be moving replicas from outside of this region into [5, 15]. But, this would reduce the exchange rate, sometimes dramatically. In my experience, this is very, very hard to converge. Our strategy will be to instead increase the number of replicas.

## Round Trips

```python
print ladder1.round_trips, ladder2.round_trips
print float(ladder1.round_trips) / ladder2.round_trips
```

The presence of the barrier has reduced the number of round trips by a factor of ~9. We would expect to get much poorer sampling in this system, even though the exchange rate is almost identically.

# Correlation times

First let's compute some data.

```python
w1 = Walker(make_energy(5.0, 100, 0.001), 50, 100, width, height)
w2 = Walker(make_energy(5.0, 400, 0.001), 50, 400, width, height)
coords1 = np.array([w1.steps(10) for i in range(1000000)])
coords2 = np.array([w2.steps(10) for i in range(1000000)])
pp.plot(coords1[:,0], coords1[:, 1])
pp.plot(coords2[:,0], coords2[:, 1])
pp.xlim(0, 200)
pp.ylim(0, 800)
```

Walker w2 experiences a large, orthogonal barrier along the x-axis, which leads to slow relaxation rates. We'll plot the autocorrelation functions and compute the autocorrelation times to get a feel for this.

```python
#
# here's a quick fft-based autocorrelation routine to replace
# np.correlate, which is dog slow
#
def acorr(inp):
    data = inp.copy()
    n = len(data)
    data = inp - np.mean(inp)
    data_pad = np.zeros(2 * n)
    data_pad[n/2:n/2+n] = data
    norm = np.sum(data**2)
    corr = signal.fftconvolve(data_pad, data[::-1], mode='valid') / norm
    return corr[(corr.size / 2):]

corr00 = acorr(coords1[:, 0])
corr10 = acorr(coords2[:, 0])
corr01 = acorr(coords1[:, 1])
corr11 = acorr(coords2[:, 1]);
```

```python
pp.plot(corr00, 'k-')
pp.plot(corr01, 'k--')
pp.plot(corr10, 'r-')
pp.plot(corr11, 'r--')
pp.xlim(0, 1000);
pp.figure()
pp.plot(corr10, 'r-')
pp.xlim(0, 20000);
```

```python
def exp_func(x, b):
    return np.exp(-b * x)

n = 20000
popt00, _ = scipy.optimize.curve_fit(exp_func, np.array(range(0, n)), corr00[:n])
popt01, _ = scipy.optimize.curve_fit(exp_func, np.array(range(0, n)), corr01[:n])
popt10, _ = scipy.optimize.curve_fit(exp_func, np.array(range(0, n)), corr10[:n])
popt11, _ = scipy.optimize.curve_fit(exp_func, np.array(range(0, n)), corr11[:n])
print 1.0 / popt00, 1.0 / popt01
print 1.0 / popt10, 1.0 / popt11
print popt00 / popt10, popt01 / popt11
```

The dynamics of w2 are slowed down by a factor of ~500 compared to w1. This is due to the presence of barrier orthogonal to the umbrella sampling / replica echange coordinate.

# With extra replicas

Now let's try it with 96 extra copies of the replicas around the barrier.

We're adding 96 extra copies of replica 11, which has an umbrella center of ~413. We're using 5x as much computer power total.

```python
centers = np.linspace(100, 700, 24)
parms = [UmbrellaParam(x, 0.001) for x in centers]

def num_copies(center):
    if center >= 400 and center <= 425:
        return 97
    else:
        return 1
    
states = [[State((50 if c < 400 else 150, c)) for i in range(num_copies(c))] for c in centers]
ladder3 = Ladder(parms, states, functools.partial(make_energy, 5.0), width, height)
```
